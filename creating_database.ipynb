```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base

engine = create_engine('sqlite:///E:\MAID\ADSA_pro\code\movielens_all.db', echo=True)
Base = declarative_base()
# creates the Base class, which is what all models(tables) inherit from and how they get SQLAlchemy ORM functionality.
Base.metadata.create_all(engine)
```

```python
movies_list = pd.read_csv('E:\MAID\ADSA_pro\code\movies.csv')
popular_movies = pd.read_csv('E:\MAID\ADSA_pro\code\popular_movies.csv')
movies_stats = pd.read_csv('E:\MAID\ADSA_pro\code\movies_stats.csv')
```

```python
movies_stats.to_sql('movies_stats',if_exists='replace',con=engine)
```

```python
movies_list.to_sql('movies_list',if_exists='replace',con=engine)
```

```python
popular_movies.to_sql('popular_movies',if_exists='replace',con=engine)
```
